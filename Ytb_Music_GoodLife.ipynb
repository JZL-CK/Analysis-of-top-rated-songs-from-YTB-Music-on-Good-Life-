# YTB Music details (web-scraping)

```python
dev = "Google API key"
```

```python
import googleapiclient.discovery
import pandas as pd

api_service_name = "youtube"
api_version = "v3"
DEVELOPER_KEY = dev

youtube = googleapiclient.discovery.build(
    api_service_name, api_version, developerKey=DEVELOPER_KEY)

music_ids = ['SmXbCBAHTfo', 'oND0Huaw5Fc', 'ss9VZ1FHxy0', 'TcpUF5bidoY', 'Onm7ywem66U', '6f8gDL-wPN8', 'U-N8gJ4HdYc', 'mDxwN4jXRQs', '8D3pVWQtxdU', '1BRTttDNvgE', 'zOFygHcpwb8', 'oM9TVbFrYCA', 'eA8gUN34nlw', 'oORFMeSua0w', '6Wer19_wCQI', 'yhyMyewwm0U', 'BNMKGYiJpvg', 'AzKbe3Ou93Q', 'NRIqzLVLvLM', 'h5vB1ffj520', 'szTT_B8WxSc', 'B-or3OFG0Rk', 'CJiDerQXyoo', 'ZOhGZVjotNY', 'ZOhGZVjotNY', '8Nd0hbEPMyc', '09h2zA0bTpI', '1xRH2T8Mgjw', 'VMWJhmYRsqk', 'suj8DEg56cc', 'eedqohjwxGk', 'fGco4n9EL_k', 'dmZhIDgQ1E0', 'ZFXKNr5JfWI', 'jKkFtoGH_f8', 'FSoINHyIaSo', 'czRi5flFEbA', 'q6VSFi4lBRU', 'niY-060vmg0', 'Asfx6CHsO3k', 'FHnjibESzSo', 'hyTeuuKLB8w', 'h3FbaVREftI', '-2vTfHMOzAk', 'r-ncxwqnEoM', 'RhIM17X_7OU', 'vkkJRMaqxN8', 'iVp5PVZj2PU', 'Q0ut_pRDsyk', 'u-s62ZdMjKE', 'zvsPAwsdzI4', 'rCZpohjvuoM']

music_data = {
    "song_name": [],
    "artist": [],
    "year_of_release": [],
    "upvotes": [],
    "playback_time": [],
    "music_link": []
}

for music_id in music_ids:
    # Request video details
    video_request = youtube.videos().list(
        part="snippet,statistics,contentDetails",
        id=music_id
    )
    video_response = video_request.execute()

    # Extract video details
    video_details = video_response['items'][0]['snippet']
    statistics = video_response['items'][0]['statistics']
    content_details = video_response['items'][0]['contentDetails']

    song_name = video_details['title']
    artist = video_details['channelTitle'].replace(" - Topic", "")
    year_of_release = video_details['publishedAt'][:4]
    upvotes = statistics.get('likeCount', 0)
    playback_time = content_details['duration']
    music_link = f"https://www.youtube.com/watch?v={music_id}"

    # Append the details to the dictionary
    music_data["song_name"].append(song_name)
    music_data["artist"].append(artist)
    music_data["year_of_release"].append(year_of_release)
    music_data["upvotes"].append(upvotes)
    music_data["playback_time"].append(playback_time)
    music_data["music_link"].append(music_link)

# Create a DataFrame from the dictionary
df_music = pd.DataFrame(music_data)

# Save the DataFrame to a CSV file
df_music.to_csv('youtube_videos_details.csv', index=False)

print("Details of the videos have been saved to 'youtube_videos_details.csv'.")
```
